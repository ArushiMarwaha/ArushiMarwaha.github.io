```python
class List:
    def __init__(self, initlist=[]):
        self.value = None
        self.next = None
        for x in initlist:
            self.append(x)
        return

    def isempty(self):
        return(self.value == None)
    
    def appendi(self, v):   # append, iterative
        if self.isempty():
            self.value = v
            return
        
        temp = self
        while temp.next != None:
            temp = temp.next

        temp.next = List()
        temp.next.value = v 
        return

    def appendr(self, v):   # append, recursive
        if self.isempty():
            self.value = v
        elif self.next == None:
            self.next = List([v])
        else:
            self.next.appendr(v)
        return

    def append(self, v):
        self.appendr(v)
        return

    def insert(self, v):
        if self.isempty():
            self.value = v
            return

        newnode = List()
        newnode.value = v
        
        # Exchange values in self and newnode
        (self.value, newnode.value) = (newnode.value, self.value)

        # Switch links
        (self.next, newnode.next) = (newnode, self.next)

        return

    def delete(self, v):   # delete, recursive
        if self.isempty():
            return

        if self.value == v:
            if self.next != None:
                self.value = self.next.value
                self.next = self.next.next
            else:
                self.value = None
            return
        else:
            if self.next != None:
                self.next.delete(v)
                # Ensure that there is no empty node at the end of the list
                if self.next.value == None:
                    self.next = None
        return
    
    def __str__(self):
        # Iteratively create a Python list from linked list
        # and convert that to a string
        selflist = []
        if self.isempty():
            return(str(selflist))

        temp = self
        selflist.append(temp.value)
        
        while temp.next != None:
            temp = temp.next
            selflist.append(temp.value)

        return(str(selflist))


    # Question 1
    """ Write a function member() such that l.member(v) returns True if the value v is present in list l
    and False otherwise."""
    
    def member(self, v):
        current = self               # start traversal from the head of the list
        while current is not None:   # loop continue until the end of the list
            if current.value == v:   # checking if current node holds the target value if it holds returning True
                return True
            
            current = current.next   # moving to next node
        return False                 # if after full traversal value is not found then returning False


    # Question 2
    """  Write a function valueat() such that l.valueat(i) returns the value at index i. This is equivalent
    to l[i] for normal Python lists. As usual, valid indices run from 0 to length(l)- 1, and backwards
    from-1 to-length(l) The function should raise an IndexError if the index is not valid. The error message should include
    the index requested."""
    
    def ll_length(self):
        """ Global helper function to return the length of a linked list."""
        length = 0                   # initializing the counter
        current = self               # starting traversal from the head of linkedlist
        while current is not None: 
            length += 1              # increasing the length counter after each node
            current = current.next   # moving to the next node
        return length                # returning the total length of the linkedlist
 

    def valueat(self, i):
        """Return the value at index i, supporting negative indices."""
        if self.isempty():
            raise IndexError(f"Index {i} is out of range as the list is empty")

        length = self.ll_length()        # calling the global helper function to get the total length

        if not (-length <= i < length):  # validating invalid index
            raise IndexError(f"Index {i} is out of range.")

        if i < 0:                        # if the index is negative then adjusting it and making it positive index
            i = length + i

        current = self                    # starting the traversal
        for _ in range(i):                # moving to the ith node
            current = current.next

        return current.value              # returning the value at index i


    # Question 3
    """ Modify the function insert() so that, for a list l, l.insert(v,i) inserts v before index i. 
    In other words, after the insert, l.valueat(i) should be v. The index i is valid if it lies in the range
    0 to length(l). When i is length(l), the effect is the same as l.append(v).
    The default value for the index i is 0. So, l.insert(v) should behave like l.insert(v,0).
    The function should raise an IndexError if the index is not valid. The error message should include
    the index requested. """
    
    def insert(self, v, i=0):
        if self.isempty():              # If list is empty, inserting at head of the linkedlist
            self.value = v
            return self
    
        length = self.ll_length()       # calling the global helper function to get the total length
        
        if i < 0:                       # if the index is negative then adjusting it and making it positive index
            i = length + i
    
        if not (0 <= i <= length):      # validate invalid index
            raise IndexError(f"Index {i} is invalid index.")
    
        if i == length:                 # If index equals length, appending at end
            self.append(v)
            return
    
        # Traversing to node before insertion point
        prev = None
        current = self                 # starting the traversal
        index = 0    
        while index < i:               # moving to the (i-1)th node
            prev = current
            current = current.next
            index += 1
    
        newnode = List()               # creating a new node
        newnode.value = v              # assigning value to the new node
    
        if i == 0:
            # inserting at head using swapping and relinking
            (self.value, newnode.value) = (newnode.value, self.value)
            (self.next, newnode.next) = (newnode, self.next)
        else:
            # Inserting in the middle
            prev.next = newnode
            newnode.next = current
    
        return
       

    # Question 4
    """ Write a function deletelast() such that l.deletelast(v) deletes the last occurrence of the value
    v from the list l.
    The function should raise a ValueError if the value to be deleted is not present in the list. The
    error message should include the value to be deleted.""" 
    
    def deletelast(self, v):
        if self.isempty():               # checking if the linkedlist is empty
            return
        current = self        
        last_occurence = None            # tracking the last node with value v
        last_occurence_prev = None       # tracking node before the last occurrence
        prev = None
        
        while current is not None:       # traversing the entire list
            if current.value == v:       # updating the last occurrence pointer
                last_occurence = current
                last_occurence_prev = prev
            prev = current
            current = current.next       # moving to next node
        
        if last_occurence == None:      # if value is not found then raising a ValueError
            raise ValueError(f"The value {v} is not present in the linked list.")
        
        if last_occurence_prev is None: # if the last occurrence is at the head
            self.value = self.next.value if self.next else None     # replacing head's value with next node's value (if exists)
            self.next = self.next.next if self.next else None       # skipping over the next node to remove the original head
        else:
            last_occurence_prev.next = last_occurence.next          # linking previous node to node after the last occurrence
               
        return


    # Question 5
    """ Write a function slice() such that l.slice(i,j) returns a new list containing the slice from
    l.valueat(i) to l.valueat(j-1).
    This is equivalent to l[i:j] for normal Python lists. The function slice() always takes two
    arguments. However, follow the usual list slicing conventions in case i < 0, j <= i or j >=
    length(l). Do not generate any errors or raise any exceptions."""
    
    def slice(self, i, j):
        length = self.ll_length()       # calling the global helper function to get the total length
        if i < 0:                       # if the index is negative then adjusting it and making it positive index
            i = length + i
        if j < 0:                       # if the index is negative then adjusting it and making it positive index
            j = length + j
        # clamping i and j within valid bounds
        i = max(0, min(i, length))
        j = max(0, min(j, length))

        if j <= i:                       # if the range is invalid then returning empty list
            return List()
                
        current = self                   # start traversal from the head
        pointer = 0
        
        while current is not None and pointer < i:   # traversing until we reach the start index
            current = current.next                    # moving to next node
            pointer += 1                              # increment in current position
        newlist = List()                              # initialise a new linked list for slice
        count = 0

        while current is not None and count < (j - i):  # copying nodes within [i, j) range
            newlist.append(current.value)              # adding current node's value to new list
            current = current.next
            count += 1                                 # updating slice element count
        return newlist                                # return the sliced list
        
            
    # Question 6
    """  Write a function rotate() such that l.rotate(k) rotates the list k times in place â€” that is,
    without creating any new nodes.
    For instance if the list l holds the sequence [1,2,3,4,5], l.rotate(1) would modify the sequence
    to [5,1,2,3,4]. After this update, l.rotate(2) would modify the sequence to [3,4,5,1,2].
    Calling l.rotate(2) one more time would then restore the original sequence."""
    
    def rotate(self, k):
        if k <0:
            raise ValueError(f"{k} is an invalid input.")
            
        length = self.ll_length()
        
        if length == 0 or k % length == 0:  # return if list is empty or rotation is multiple of length
            return 
    
        k = k % length  # normalize k if greater than length
        
        # performing rotation k times (each time move last node's value to the head)    
        for i in range(k):
            tail=self  
            
            while tail.next!=None: #traversing to the last node
                tail=tail.next
                
            prev_value, current =tail.value,self  #storing the last node's value and start from the head          
            
            while current!=None:  #shifting all node values one position to the right
                
                temp = current.value      # temporarily hold current node's value
                current.value=prev_value  # replacing it with the previous node's value
                current=current.next
                prev_value=temp           #updating prev_value for next iteration
                
        return self    # returning the rotated list 
        
        
        
```
