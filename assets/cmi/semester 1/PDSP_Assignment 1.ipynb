### Question 1  

**Which team had the highest winning percentage?**

Write a function `max_win_percentage(m)` that takes a list of matches and returns the team that won the largest fraction of matches that it played.



````python
#Helper Function

def match_result(ml):
    # Dictionary to store results for each team
    # Structure: { 'TeamName': {'matchesplayed': x, 'win': y, 'lost': z} }
    team_result={}
    
    for match in ml:
        
        # Extract team names and relevant match details from the tuple
        team1 = match[1]                 # first team in the match
        team2 = match[2]                 # second team in the match
        twinner = match[3]               # winner of the match
        toss_decision = match[4]         # toss decision: "Bat" or "Bowl"
        win_by_runs = match[5]           # non-empty if match decided by runs, empty otherwise
        win_by_wickets = match[6]        # non-empty if match decided by wickets, empty otherwise

        # initialize stats for both teams 
        if team1 not in team_result:  
            team_result[team1] = {'matchesplayed':0, 'win': 0, 'lost': 0}
            
        if team2 not in team_result:
            team_result[team2] = {'matchesplayed':0, 'win': 0, 'lost': 0}

        # increment the number of matches played for both teams
        team_result[team1]['matchesplayed'] = team_result[team1]['matchesplayed'] + 1   
        team_result[team2]['matchesplayed'] = team_result[team2]['matchesplayed'] + 1       
        
        
        # update win/loss counts based on how the match was decided and toss result
        #
        # Cases to handle:
        # Case 1. Toss decision = "Bat" and match decided by wickets
        # Case 2. Toss decision = "Bat" and match decided by runs
        # Case 3. Toss decision = "Bowl" and match decided by wickets
        # Case 4. Toss decision = "Bowl" and match decided by runs
        #
        # For each scenario, check if team1 or team2 is the winner
        # and increment 'win' for the winner, 'lost' for the loser.
       

        # Case 1: 
        if win_by_runs == "" and toss_decision == 'Bat':
            if twinner == team1:   # If team1 won
                team_result[team2]['win'] += 1   # Record incorrect win for team2 
                team_result[team1]['lost'] += 1  # Record loss for team1
            else:                  # If team2 won
                team_result[team1]['win'] += 1
                team_result[team2]['lost'] += 1

        # Case 2: 
        if win_by_wickets == "" and toss_decision == 'Bowl':
            if twinner == team1:   
                team_result[team2]['win'] += 1
                team_result[team1]['lost'] += 1
            else:
                team_result[team1]['win'] += 1
                team_result[team2]['lost'] += 1

        # Case 3: 
        if win_by_runs == "" and toss_decision == 'Bowl':
            if twinner == team1:
                team_result[team1]['win'] += 1
                team_result[team2]['lost'] += 1
            else:
                team_result[team2]['win'] += 1
                team_result[team1]['lost'] += 1

        # Case 4: 
        if win_by_wickets == "" and toss_decision =='Bat':
            if twinner == team1:
                team_result[team1]['win'] += 1
                team_result[team2]['lost'] += 1
            else:
                team_result[team2]['win'] += 1
                team_result[team1]['lost'] += 1
            
        # Note: No 'else' needed, since every match is assumed to produce a winner
    
   
    return team_result
    

# Example 
match_result(matchlist)
````

````python
def max_win_percentage(m):
    # getting match statistics for all teams using match_result()
    team_stats = match_result(m)   
    
    team_win_per = {}    # Dictionary to store calculated win percentage for each team

    for team in team_stats.keys():
        # calculate win percentage:
        
        winpercent = (team_stats[team]['win'] / team_stats[team]['matchesplayed']) * 100     # (number of wins / total matches played) * 100
        
        # storing the win percentage in the dictionary
        team_win_per[team] = winpercent

    # finding the maximum win percentage
    max_win_percentage = max(team_win_per.values())
    
    # identify which team(s) have this maximum win percentage
    max_win_team = [team for team, wp in team_win_per.items() if wp == max_win_percentage]
    
    return max_win_team #, max_win_percentage

# Print each team's win percentage
team_stats = match_result(matchlist)
for team in team_stats.keys():
    winpercent = (team_stats[team]['win'] / team_stats[team]['matchesplayed']) * 100
    print(f"{team}: {winpercent:.2f}% win rate")

# Example 
result = max_win_percentage(matchlist)
print("\nTeam(s) with maximum win percentage:", result)

````

### Question 2  

**Is it better to bowl first or bat first?**  

In this question, we want to analyze match outcomes based on the team’s decision to **bowl first**. 

````python
def bowl_first_win(m):
    bowl_win = 0    #initializing the bowling team win count as zero
    total_bowl = 0  #initializing the total matches count as zero
    
    for match in m:
        win_by_runs = match[5]    
        win_by_wickets = match[6]
        
        if win_by_runs == "" :   #checking if the win by runs is empty or not
            bowl_win += 1        # if win by run variable is empty => the bowling team won hence increasing the bowling team win count
        total_bowl += 1          # increasing the total matches played count
    
    return bowl_win / total_bowl if total_bowl > 0 else 0   # returing the fraction and if the total matches played is zero will return zero

bowl_first_win(matchlist)        #trial 

# Example 
fraction = bowl_first_win(matchlist)

print(f"Bowling first win ratio: {fraction:.10f}")
````

### Question 3 
**Which team(s) batted first most often?**  

Write a function `bat_first_max(m)` that takes a list of matches and returns the team(s) that batted first most often, as a list.


````python
def bat_first_max(m):
    # dictionary to store how many times each team batted first.    Structure: { 'TeamName': {'bat_count': x} }
    team_name = {}        
    
    # iterate over each match in the match list
    for match in m:
        team1 = match[1]        
        team2 = match[2]        
        tosswinner = match[3]   # team that won the toss
        tossd = match[4]        # toss decision: "Bat" or "Bowl"
        
        if team1 not in team_name:
            team_name[team1] = {'bat_count': 0}
        if team2 not in team_name:
            team_name[team2] = {'bat_count': 0}

       
        # Determine which team batted first:

        if tossd == 'Bat':
            bfirst = tosswinner  
        else:
            # If tosswinner bowls, the opponent bats first
            bfirst = team1 if tosswinner == team2 else team2   

        # increment bat_count for whichever team batted first
        team_name[bfirst]['bat_count'] += 1

    # find the maximum number of times any team has batted first and  max_bat stores that maximum value

    max_bat = max(team['bat_count'] for team in team_name.values())   
    
    # Collect all teams that have this maximum bat_count
    result = [team for team, data in team_name.items() if data['bat_count'] == max_bat]
    
    # return the list of team(s) that batted first the most
    return result


# Example 
result = bat_first_max(matchlist)

print("Team(s) that batted first the most:", ", ".join(result))

````

### Question 4  

**If the match goes into the final over (number of balls left is less than 6), which team has an advantage — the team that batted first or the team that batted second?**  

Write a function `last_over_favours(m)` that takes a list of matches and returns:  
- `1` if the team batting first has an advantage  
- `2` if the team batting second has an advantage  


````python
#helper function
def matches_lastover(m):
    newml = []
    for match in m:
        # Check if the match went into the final over
        if match[7] < 7:
            newml.append(match)
    return newml

# get all matches that went into the last over
matches_lastover(matchlist)


def last_over_favours(m):
    # Filter matches that went to the last over
    last_over_matches = matches_lastover(m)
    bat_first_wins = 0
    bat_second_wins = 0
    
    for match in last_over_matches:
        toss_winner = match[3]
        toss_decision = match[4]
        win_by_runs = match[5]
        win_by_wickets = match[6]
        
        # figure out which team batted first and which batted second
        if toss_decision == 'Bat':
            bat_first = toss_winner
            bat_second = match[2] if toss_winner == match[1] else match[1]
        else:  # toss_decision == 'Bowl'
            bat_first = match[1] if toss_winner == match[2] else match[2]
            bat_second = toss_winner
        
        # update win counts depending on how the match ended
        if win_by_runs != "":   # if decided by runs → batting-first team won
            bat_first_wins += 1
        elif win_by_wickets != "":  # if decided by wickets → batting-second team won
            bat_second_wins += 1

    # Compare which side had more wins in last-over matches
    if bat_first_wins > bat_second_wins:
        return 1   # batting first has advantage
    else:
        return 2   # batting second has advantage


# example
last_over_favours(matchlist)
````

### Question 5  

**Which venue(s) most favour the team batting second?**  

Write a function `best_chasing_venue(m)` that takes a list of matches and  
returns the venue(s) that most favour the team batting second, as a list.  


````python
# helper function
def venue_d(m):
    venues_d = {}
    for match in m:
        venue = match[0]    # extract the venue of the match
        if venue not in venues_d:
            venues_d[venue] = 0  # initialize to 0 for counting (placeholder)
    return venues_d
    
def chasing_venue_dict(m):
    venue_dic = venue_d(m)   # get all venues initialized
    venue_favour = {}
    
    # For each venue, create a dictionary to track total matches 
    # and number of wins by batting second (chasing wins)
    for venue in venue_dic:
        venue_dic[venue] = {'total_m': 0, 'bat_2nd_win': 0}
        
    # Iterate over matches and update counts
    for match in m:
        venue = match[0]
        team1 = match[1]
        team2 = match[2]
        tosswinner = match[3]
        tossd = match[4]
        win_by_runs = match[5]
        win_by_wickets = match[6]
        
        # Increment total matches at that venue
        venue_dic[venue]['total_m'] = venue_dic[venue]['total_m'] + 1
        
        # If match was won by wickets → batting second team won (chasing win)
        if win_by_wickets != "":
            venue_dic[venue]['bat_2nd_win'] += 1
    
    return venue_dic

# Example call to get venue-wise chasing performance
chasing_venue_dict(matchlist)
````

````python
def best_chasing_venue(m):
    venue_dic = venue_d(m)   # get all venues initialized
    venue_favour = {}
    
    # For each venue, create counters for total matches and batting 2nd wins
    for venue in venue_dic:
        venue_dic[venue] = {'total_m': 0, 'bat_2nd_win': 0}
        
    # Traverse each match and update venue stats
    for match in m:
        venue = match[0]
        team1 = match[1]
        team2 = match[2]
        tosswinner = match[3]
        tossd = match[4]
        win_by_runs = match[5]
        win_by_wickets = match[6]
        
        # Increment total matches at that venue
        venue_dic[venue]['total_m'] = venue_dic[venue]['total_m'] + 1
        
        # If match was won by wickets → batting second (chasing) team won
        if win_by_wickets != "":
            venue_dic[venue]['bat_2nd_win'] += 1
            
    # Calculate chasing win fraction for each venue
    for venue, stats in venue_dic.items():
        if stats['total_m'] > 0:   # to avoid division by zero 
            favour_fraction = (stats['bat_2nd_win'] / stats['total_m'])
            venue_favour[venue] = favour_fraction
    
    # maximum chasing fraction among venues
    max_favour = max(venue_favour.values())
    
    # Collect all venues with that maximum fraction 
    max_favour_team = [venue for venue, w in venue_favour.items() if w == max_favour]   

    # Return both best venue(s) and corresponding chasing fraction
    return (max_favour_team, max_favour)
        
        
# trial 
best_chasing_venue(matchlist)

````
