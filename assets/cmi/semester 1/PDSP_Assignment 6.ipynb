## Method 1: Function Definition Without Using Dynamic Programming


```python
def LongestPalindrome(N, word):
    if N == 0:
        return 0, ""
        
    # initialising longest palindrome and its length
    pali = word[0]
    maxlen = 1
    wordlist = list(word)   

    # -------------------------------------------- checking odd length palindromes 
    for mid in range(N):
        left, right = mid, mid
        
        # expanding around the center while matching characters
        while left >= 0 and right < N and word[left] == word[right]:
            curlen = right - left + 1
            curword = word[left:right+1]
            
            # Updating longer palindrome

            if curlen > maxlen:
                maxlen = curlen
                pali = curword
                
            # Updating lexicographically smaller palindrome

            elif curlen == maxlen and curword < pali:
                pali = curword

            left -= 1   # expanding left
            right += 1  # expanding right

    # -------------------------------------------- checking even length palindromes
    for mid in range(N):
        left, right = mid, mid + 1
        
        # expanding around the center while matching characters
        while left >= 0 and right < N and word[left] == word[right]:
            curlen = right - left + 1
            curword = word[left:right+1]
            # Updating longer palindrome
            if curlen > maxlen:
                maxlen = curlen
                pali = curword
                
            # Updating lexicographically smaller palindrome

            elif curlen == maxlen and curword < pali:
                pali = curword

            left -= 1   # expanding left
            right += 1  # expanding right
                
    return maxlen, pali
```

## Method 2: Function Definition Using Dynamic Programming

```python
def DPLongestPalindrome(N, word):
    if N == 0:
        return 0, ""
        
    # creating DP table for storing palindrome states
    dp = [[False] * N for _ in range(N)]
    
    # Initialising the longest palindrome and its length

    pali = word[0]
    maxlen = 1

    for i in range(N):     # marking all single characters as palindromes
        dp[i][i] = True

    for i in range(N - 1):     # checking all two-character palindromes
        if word[i] == word[i + 1]:
            dp[i][i + 1] = True
            
            if 2 > maxlen:    # updating longest palindrome of length 2
                maxlen = 2
                pali = word[i:i+2]
                
            # updating lexicographically smaller palindrome
            elif 2 == maxlen and word[i:i+2] < pali: 
                pali = word[i:i+2]
                
    # checking palindromes of length 3 and above
    for length in range(3, N + 1):
        
        for i in range(N - length + 1):
            j = i + length - 1  # setting ending index
            
            # checking palindrome using DP recurrence
            if word[i] == word[j] and dp[i + 1][j - 1]:
                dp[i][j] = True
                curword = word[i:j+1]
                
                # updating longer palindrome
                if length > maxlen:
                    maxlen = length
                    pali = curword
                    
                # updating lexicographically smaller palindrome
                elif length == maxlen and curword < pali:
                    pali = curword

    return maxlen, pali
```
